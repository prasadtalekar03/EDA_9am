```python
# Searech on youtube how to do KNNImputer, It is not teach by sir very well

What is KNN Imputer?
KNN Imputer in Machine Learning (K-Nearest Neighbors) KNN Imputer is a machine learning algorithm 
that can help you fill in the blanks in your data.
It does this by finding the most similar data points to the one with the missing value 
and then using the average of those values to fill in the missing value.
```

**KNN Immputer**

- K - means any number

- NN - So NN means Nearest Neighbours

-  Where k = hyper parameter which means user can change

- In machine learing there are two things

- 1) parameter  2) Hyper parameter

- 1) parameter means model will update the value internally
 
  2) hyper parameter means user will change it by their will
 
            def(a,b):
               return(a+b)
                    
            add(10,20)
            
         # Here 10,20 is hyper parameter and if we want to change we can change it        

 
-  Idea: Instead of taking all the sample average

      - first fix the neighbours number i.e. k
 
      - Then find the neighbours using distance matric
 
      - Then take the average of those samples to fill the missing values
