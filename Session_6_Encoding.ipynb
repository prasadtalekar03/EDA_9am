**LabelEncoder Vs OneHotEncoder**

- Generally we don't want relation between input columns

- we need both input columns are **independent(X,Y)** each other

- we need both input column are perpendicular each other

- we need both input columns are **orthogonal** each other 

- we need both input columns are **90 degree phase shift** each other

- OneHotEncoder will acheive above

- But OneHotEncoder creates new columns

    - columns are increase means dimenssion are increase

    - More Dimension data need more procssing time

    - more dimension data need more infra

    - This is called **curse of Dimensionally**
