```python
df.drop('age')    
# label = either 0 or Names
# index= 0  1  2
# column = Names  Age  City  age
# here python will assume age is a label
# and python will ask qn to himself: is it index label or column label
# for that we need to provide axis 
# axis= 1 means for cloumn
# axis= 0 means for rows
# by default python will take axis=0 
# it means python ('age') ko index samjhega,instead of column
# beacuse when we do shift+tab then, axis: 'Axis' = 0
# beacuse python will assume age is a index label and
# but actually me age is column
# so if we run this it will give error
```

```python
  df.drop('age',axis=0)   
# this is also error
# beacuse axis = 0 means index label
# so to avoid this error use axis=1 
# and we already know that axis=1 is used for column label
```

```python
# thats why you will have to right like this
df.drop('age',
        axis=1)
# now the column of age will override
```

```python
df   # so here age [30,32,34] colum will override 
     # but we are able to see old column still
     # beacuse we did not apply inplace=true
```

```python
df.drop('age',
        axis=1,
       inplace=True)
# inplace will help you to override the age column
```

**Step-12: how to append(add) a row**

In append there is loc & iloc function

- loc

- iloc

```python
# df1.loc[rows,colums]
# here the left side of comma is specifically for rows & right side are specifically for colums
# we can select a particular row and column via using a list
# and we already know how to do that
# list: [start:stop:step]
```

```python
# for eg.

df1.loc[[3:7],[1:3]]
# so here are 4 rows avaliable from 3 to 7
# and 3 columns from 1 to 3 
# now if we run this it will show error 
# it means loc function is not accepting this kind of list representation
```

```python
# so do like this it will not show error
df1.loc[3:7]
# but it is only showing rows representation 
# so if I want to see column representation use iloc function (df1.iloc[3:7,1:3]) or (df1.loc[3:7,['square_value','cube_value']])
```
