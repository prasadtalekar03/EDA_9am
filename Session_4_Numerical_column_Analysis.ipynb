**Now let's prove Emperical rule**

**Emperical rule**

- u-1*sigma   to   u+1sigma  = 68% of data should coverage 

 - so this equation is true equation and I want to prove it let's prove it

- To prove use this steps

    - step-1: wage data mean = u
    - step-2: wage data std = (sigma)
    - step-3: LB=u-1*sigma,           LB=lower bound
    - step-4: UB=u+1*sigma,           UB=upper bound
    - step-5: con1=visa_df['prevailing_wage']<LB
    - steo-6: con2=visa_df['prevailing_wage']<UB
    - step-7: con1==con2 &con2
    - step-8: len(visa_df[con])


```python
          Q1-1.5IQR   Q1   median  Q3   Q3+1.5IQR
                       |-----:-----|        
       o      |--------|     :     |--------|    o  o
                       |-----:-----|
     flier             <----------->            fliers
```

```python
wage_data=visa_df['prevailing_wage']
U=wage_data.mean()
sigma=wage_data.std()
LB=U-1*sigma
UB=U+1*sigma
con1=visa_df['prevailing_wage']>LB
con2=visa_df['prevailing_wage']<UB
con=con1 & con2
len(visa_df[con])


# 17171==> so this is 68% of data which we have to proved
# so this 68% of data in b/w u-1sigma to u+1sigma
```
